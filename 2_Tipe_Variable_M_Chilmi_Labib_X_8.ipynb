```python
#Perkalian

jumlah_kardus_air_minum = 3
jumlah_botol_perkardus = 10
total_botol_air_minum = jumlah_kardus_air_minum * jumlah_botol_perkardus

print(total_botol_air_minum)
```

```python
#Pembagian

total_botol_air_minum = 30
jumlah_anak_yatim = 7
jumlah_botol_tiap_anak_yatim = total_botol_air_minum / jumlah_anak_yatim

print(jumlah_botol_tiap_anak_yatim)
```

```python
#Pembagian bulat

total_botol_air_minum = 30
jumlah_anak_yatim = 7
jumlah_botol_tiap_anak_yatim = total_botol_air_minum // jumlah_anak_yatim

print(jumlah_botol_tiap_anak_yatim)
```

```python
#Sisa bagi

total_botol_air_minum = 30
jumlah_anak_yatim = 7
sisa_bagi = total_botol_air_minum % jumlah_anak_yatim

print(sisa_bagi)
```

Latihan Soal Aritmatika :
1. Penjumlahan :
Dina memiliki 8 apel. ibunya memberi 5 lagi apel.
Berapa jumlah apel Dina sekarang ?
2. Pengurangan :
Seorang petani memetik 30 mangga. Lalu 12 mangga sudah matang dan dijual.
Berapa mangga yang tersisa ?
3. Perkalian :
Sebuah kelas memiliki 6 meja. Setiap meja ditempati 4 siswa.
Berapa jumlah siswa seluruh nya ?
4. Pembagian :
Ibu membeli 30 roti untuk dibagikan kepada 4 orang secara merata.
Berapa roti yang diterima setiap orang ?  
5. Sisa bagi :
Ayah membeli 25 donat. Donat tersebut akan dibagikan ke 6 anak sama rata.
Berapa donat yang tersisa ?

6.Perpangkatan ( luas persegi ) :
Sebuah taman berbentuk persegi dengan panjang sisi 7 meter.
Berapa luas taman tersebut?

```python
#Penjumlahan

apel_dina = 8
apel_ibu = 5

total_apel = apel_dina + apel_ibu

print(total_apel)
```

```python
#Pengurangan

mangga_total = 30
mangga_terjual = 12

mangga_sisa = mangga_total - mangga_terjual

print(mangga_sisa)
```

```python
#Perkalian

meja = 6
siswa_per_meja = 4

jumlah_siswa = meja * siswa_per_meja

print(jumlah_siswa)
```

```python
# Pembagian

roti_total = 30
orang = 4

roti_per_orang = roti_total / orang

print(roti_per_orang)
```

```python
#Sisa bagi

donat_total = 25
anak = 6

sisa_bagi = donat_total % anak

print(sisa_bagi)
```

```python
#Perpangkatan (luas_persegi)

sisi = 7
luas_taman = sisi ** 2

print(luas_taman)
```
